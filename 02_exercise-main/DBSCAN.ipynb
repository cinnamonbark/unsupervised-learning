We add the points into our points list 

```python
import pandas as pd
from itertools import product
import numpy as np


# Define the data points
points = {
    'A': (1, 1), 'B': (2, 1), 'C': (1, 2), 'D': (2, 2), 'E': (3, 5),
    'F': (3, 9), 'G': (3, 10), 'H': (4, 10), 'I': (4, 11), 'J': (5, 10),
    'K': (7, 10), 'L': (10, 9), 'M': (10, 6), 'N': (9, 5),
    'O': (10, 5), 'P': (11, 5), 'Q': (9, 4), 'R': (10, 4),
    'S': (11, 4), 'T': (10, 3)
}



print(len(points))
```

Transform the list into an array 

```python
X = np.array(list(points.values()))

print(X)
```

 Perform the DBSCAN with different hyperparamethers, print the results 

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# parameter setting as a list of tuples 
P=[(1.1, 2), (1.1, 3), (1.1, 4), (2.1, 4), (4.1, 5), (4.1, 4)]

#loop through each tuple
for eps, minPts in P:
    db = DBSCAN(eps,min_samples=minPts,  metric='manhattan')
    print(db.fit_predict(X)) #different numbers mean different clusters and -1 means noise 
    plt.scatter(X[:,0], X[:,1], c=db.fit_predict(X)) #c stands for color 
    plt.title("DBSCAN for eps=%s and minPts=%s" % (eps, minPts))
    plt.show()

```

We can see by the graphs that the best paramethers from those we've chosen are: eps 2.1 and minPoints 4, or 1.1 and 3. This depends on considering some of the points to behave as a "noise" or include them to the clustering
